## CASE STUDY:  ORDINARY DIFFERENTIAL EQUATIONS
### NUMERICAL COMPUTING

__________

#### Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
```

_________
#### Code:

```python
r = 0.3
K = 10000
h = 1
P0 = 500
days = 20

P = [P0]
t = [0]

for i in range(1, days + 1):
    P_prev = P[-1]
    dPdt = r * P_prev * (1 - P_prev / K)
    P_next = P_prev + h * dPdt
    P.append(P_next)
    t.append(i)

data = {'Day': t, 'Estimated Users': P}
df = pd.DataFrame(data)

print(df.to_string(index=False))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, P, marker='o', linestyle='-', color='blue', label='Estimated Users')
plt.axhline(y=K, color='red', linestyle='--', label='Market Saturation (K=10,000)')
plt.title('User Growth Simulation using Eulers Method')
plt.xlabel('Time (days)')
plt.ylabel('Number of Users')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
